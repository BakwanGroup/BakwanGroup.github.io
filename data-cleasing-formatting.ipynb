```python
# CELANSING

import xml.etree.ElementTree as ET
from xml.etree.ElementTree import Element, SubElement

# begin
tree1 = ET.parse('api.all.xml')
root1 = tree1.getroot()
     
# hapus yg tak perlu
for plc in root1.findall('item'):
    check = plc.find('openingHours')
    if not check is None:  
        for opening in plc.findall('openingHours'):
            opening.clear()
            plc.remove(plc.find('openingHours'))
            
    check = plc.find('additionalInfo')
    if not check is None:  
        for opening in plc.findall('additionalInfo'):
            opening.clear()
            plc.remove(plc.find('additionalInfo'))

        
tree1.write('api.all.xml') 
```

```python
# COMBAINE AND FORMATTING ALL XML

import xml.etree.ElementTree as ET
from xml.etree.ElementTree import Element, SubElement

# source xml formatter: https://blog.csdn.net/shinobiii/article/details/8253976
def prettyXml(element, indent, newline, level = 0):
    if element:  
    # 判断element是否有子元素  
        if element.text == None or element.text.isspace(): 
        # 如果element的text没有内容  
            element.text = newline + indent * (level + 1)    
        else:  
            element.text = newline + indent * (level + 1) + element.text.strip() + newline + indent * (level + 1)  
        #else:  # 此处两行如果把注释去掉，Element的text也会另起一行  
        #element.text = newline + indent * (level + 1) + element.text.strip() + newline + indent * level  
    temp = list(element) # 将elemnt转成list  
    for subelement in temp:  
        if temp.index(subelement) < (len(temp) - 1): # 如果不是list的最后一个元素，说明下一个行是同级别元素的起始，缩进应一致  
            subelement.tail = newline + indent * (level + 1)  
        else:  # 如果是list的最后一个元素， 说明下一行是母元素的结束，缩进应该少一个  
            subelement.tail = newline + indent * level  
        prettyXml(subelement, indent, newline, level = level + 1) # 对子元素进行递归操作  

# begin
tree1 = ET.parse('anotation.all.xml')
tree2 = ET.parse('wiki.all.xml')
tree3 = ET.parse('gmaps.all.xml')

root1 = tree1.getroot()
root2 = tree2.getroot()
root3 = tree3.getroot()

# cek kesamaan nama               
for i, phn1 in enumerate(root1.findall('place')):
    # Ambil dari wiki
    for j, phn2 in enumerate(root2.findall('Dest')):
        if (phn1.find('name').text == phn2.find('name').text):
            # Buat node loc
            loc = SubElement(phn1, 'loc')
            
            # Anak node loc
            prov = SubElement(loc, 'prov')
            kabkot = SubElement(loc, 'kabkot')
                            
            # Ambil data dari wiki, masukkin ke anot
            temp = phn2.find('prov').text
            prov.text = temp
            temp = phn2.find('kabkot').text
            kabkot.text = temp
            
            # Ambil dari api
            for j, phn3 in enumerate(root3.findall('item')):
                if (phn1.find('name').text == phn3.find('searchString').text):
                    # Anak node loc
                    address = SubElement(loc, 'address')
                    url = SubElement(loc, 'url')
                    website = SubElement(phn1, 'website')
                    phone = SubElement(phn1, 'phone')
                    totalScore = SubElement(phn1, 'totalScore')
                    reviewDistribution = SubElement(phn1, 'reviewDistribution')
                    
                    # Ambil data dari api, masukkin ke anot
                    temp = phn3.find('address').text
                    address.text = temp
                    temp = phn3.find('url').text
                    url.text = temp
                    temp = phn3.find('website').text
                    website.text = temp
                    temp = phn3.find('phone').text
                    phone.text = temp
                    temp = phn3.find('totalScore').text
                    totalScore.text = temp
                                    
                    # Make nodeChild review
                    oneStar = SubElement(reviewDistribution, 'oneStar')
                    twoStar = SubElement(reviewDistribution, 'twoStar')
                    threeStar = SubElement(reviewDistribution, 'threeStar')
                    fourStar = SubElement(reviewDistribution, 'fourStar')
                    fiveStar = SubElement(reviewDistribution, 'fiveStar')
                    
                    # Input review 
                    if len(phn3.find('reviewsDistribution')):
                        temp = phn3.find('reviewsDistribution').find('oneStar').text                
                        oneStar.text = temp
                        temp = phn3.find('reviewsDistribution').find('twoStar').text
                        twoStar.text = temp
                        temp = phn3.find('reviewsDistribution').find('threeStar').text
                        threeStar.text = temp
                        temp = phn3.find('reviewsDistribution').find('fourStar').text
                        fourStar.text = temp
                        temp = phn3.find('reviewsDistribution').find('fiveStar').text
                        fiveStar.text = temp
                    
prettyXml(root1, '\t', '\n')       
tree1.write('final.all.xml')    
```
